```python
# Этот код создает квантовую схему с 4 кубитами и 2 классическими битами. 
# Затем на первом и втором кубите применяются операции X (NOT), 
# чтобы установить их в состояние |1⟩. Затем выполняются две операции CNOT, 
# чтобы скопировать состояния первого и второго кубитов на третий кубит. 
# Наконец, выполняются две операции измерения на 3-м и 4-м кубитах и их результаты записываются в 0-й и 1-й классические биты. 
# Когда схема будет запущена на квантовом компьютере, 
# она будет измерять два кубита и записывать результаты измерений на два классических бита.
qc = QuantumCircuit(4,2)
qc.x(0)
qc.x(1)
qc.cx(0,2)
qc.cx(1,2)
qc.measure(2,0)
qc.measure(3,1)
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(4,2)
qc.x(0)
qc.x(1)
qc.cx(0,2)
qc.cx(1,2)
qc.ccx(0,1,3) # Выполняется операция Toffoli (CCNOT), 
# чтобы применить операцию NOT на четвертом кубите, когда первые два кубита имеют значение 1.
qc.measure(2,0)
qc.measure(3,1)
qc.draw(output='mpl')
```

```python
counts = execute(qc, BasicAer.get_backend('qasm_simulator')).result().get_counts(qc)
print(counts)
plot_histogram(counts)

# Получаем стопроцентную вероятность значения {10}
```
