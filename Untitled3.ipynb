```python
from qiskit import *
```

```python
import matplotlib.pyplot as plt
```

```python
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(2)
```

```python
qc.cx(0,1)
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(4,2)
qc.x(0)
qc.x(1)
qc.cx(0,2)
qc.cx(1,2)
qc.measure(2,0)
qc.measure(3,1)
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(4,2)
qc.x(0)
qc.x(1)
qc.cx(0,2)
qc.cx(1,2)
qc.ccx(0,1,3)
qc.measure(2,0)
qc.measure(3,1)
qc.draw(output='mpl')
```

```python
counts = execute(qc,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
